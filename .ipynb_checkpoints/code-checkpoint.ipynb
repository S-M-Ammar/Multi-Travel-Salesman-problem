```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import random
```

```python
file = open('./dj38.tsp', 'r')
Lines = file.readlines()
```

```python
Lines = Lines[10:]
df = pd.DataFrame(columns=['Longitude','Lattitude'])
x_cord = []
y_cord = []
index = 0
for line in Lines:
    line = line.replace("\n","")
    raw_elements = line.split(" ")
    x_cord.append(raw_elements[1]) # longitude
    y_cord.append(raw_elements[2]) # lattitude
    index+=1
```

```python
df['Longitude'] = x_cord
df['Lattitude'] = y_cord
```

```python
df
```

```python
from sklearn.cluster import KMeans
distortions = []
K = range(2,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df)
    distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

#### Choosing number of clusters => k = 3


```python
k=3
kmeanModel = KMeans(n_clusters=k,init='k-means++')
kmeanModel.fit(df)
kmeanModel.cluster_centers_
```

```python
center_0 = kmeanModel.cluster_centers_[0]
center_1 = kmeanModel.cluster_centers_[1]
center_2 = kmeanModel.cluster_centers_[2]
```

```python
df['Cluster_Number'] = kmeanModel.labels_.tolist()
```
